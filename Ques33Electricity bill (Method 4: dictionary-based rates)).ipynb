```python
units = int(input("Enter units: "))

rates = {1: 5, 2: 7, 3: 10}

if units <= 100:
    bill = units * rates[1]
elif units <= 200:
    bill = units * rates[2]
else:
    bill = units * rates[3]

print("Total bill amount:", bill)
```
