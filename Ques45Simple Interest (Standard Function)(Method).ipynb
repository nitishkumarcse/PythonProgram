```python
def get_si(p, r, t):
    interest = (p * r * t) / 100
    return interest

p_val = float(input("Enter P: "))
r_val = float(input("Enter R: "))
t_val = float(input("Enter T: "))

print("Simple Interest:", get_si(p_val, r_val, t_val))
```
