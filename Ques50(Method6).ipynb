```python
# Lambda version of SI formula
calc_interest = lambda p, r, t: (p * r * t) / 100

p = float(input("Enter P: "))
r = float(input("Enter R: "))
t = float(input("Enter T: "))

result = calc_interest(p, r, t)
print(f"Simple Interest (Lambda): {result}")
```
