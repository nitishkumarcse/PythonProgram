```python
units = float(input("Enter units consumed: "))

if units <= 100:
    total_bill = units * 5  # $5 per unit
elif units <= 300:
    total_bill = (100 * 5) + (units - 100) * 7  # $7 for next 200 units
else:
    total_bill = (100 * 5) + (200 * 7) + (units - 300) * 10 # $10 for remaining

print(f"Total bill amount: {total_bill:.2f}")
```
